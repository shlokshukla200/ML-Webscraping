```python
import requests  # Helps us get data from websites
from bs4 import BeautifulSoup  # Helps us read and pull text from webpages
import os # Helps with file path operations

def scrape_study_content(url, output_file):
  try:
    # Pretend to be a web browser so the website lets us in
    headers = {
        'User-Agent': 'Mozilla/5.0'
    }
    response = requests.get(url, headers=headers)  # Visit the webpage
    response.raise_for_status()  # Check if the visit worked

    # Turn the webpage into something we can read easily
    soup = BeautifulSoup(response.text, 'html.parser')

    # Find the title (Usually in a big heading tag called <h1>)
    title_element = soup.find('h1')
    title_text = title_element.get_text(strip=True) if title_element else "No Title Found"

    # Find all paragraphs (<p>) to get the main content
    content = []
    for paragraph in soup.find_all('p'):
      text = paragraph.get_text(strip=True)  # Get clean text without extra spaces
      if text:
        content.append(text)

    # Save the title and content to a file
    with open(output_file, 'w', encoding='utf-8') as file:
      file.write(f"Title: {title_text}\n\n")  # Write the title
      file.write("Content:\n")  # Write a label for content
      for line in content:
        file.write(line + "\n")

      print(f"Saved content to {output_file}")
      # Print the absolute path of the saved file
      print(f"File saved at: {os.path.abspath(output_file)}")


  except Exception as e:
    print(f"Oops, something went wrong: {e}")

# --- Example: Try it with a webpage ---
url = 'https://www.geeksforgeeks.org/machine-learning/machine-learning-algorithms/'  # Example webpage
output_file = 'study_content.txt'  # Where we'll save the content

# Run the scraper
scrape_study_content(url, output_file)

# Read and display the content of the saved file
print("\n--- Content of the output file ---")
try:
    with open(output_file, 'r', encoding='utf-8') as file:
        print(file.read())
except FileNotFoundError:
    print(f"Error: The file {output_file} was not found.")
except Exception as e:
    print(f"An error occurred while reading the file: {e}")
```
